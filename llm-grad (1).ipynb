```python
from docx import Document

# Load the Word document from the specified path
file_path = '/kaggle/input/datasetgrad/dataset Grad.docx'
doc = Document(file_path)

# Extract questions from the document
questions = [para.text for para in doc.paragraphs if para.text.strip()]
print(f"Extracted {len(questions)} questions.")

# Display the first few questions
for i, question in enumerate(questions[:5], start=1):
    print(f"Question {i}: {question}")
```

```python
import re

# Function to generate Cypher queries from questions
def generate_cypher(question):
    question = question.lower().strip()

    # Handling 'What is the prerequisite for' questions
    if "prerequisite for" in question:
        course = question.split("prerequisite for")[-1].strip().replace("?", "")
        return f"MATCH (c:Course {{name: '{course}'}})-[:PRE_REQUISITE_FOR]->(prereq:Course) RETURN prereq.name AS Prerequisite"
    
    # Handling 'What courses are unlocked' or 'open by' questions
    elif "courses unlocked" in question or "open by" in question:
        course = question.split("after completing")[-1].strip().replace("?", "").replace("open by", "").strip()
        return f"MATCH (c:Course {{name: '{course}'}})-[:UNLOCKS]->(unlocked:Course) RETURN unlocked.name AS UnlockedCourses"
    
    # Handling 'mandatory' courses question
    elif "mandatory" in question or "required" in question:
        return "MATCH (c:Course {mandatory: true}) RETURN c.name AS MandatoryCourses"
    
    # Handling 'technical electives' question
    elif "technical electives" in question:
        return "MATCH (e:Elective {type: 'technical'}) RETURN e.name AS TechnicalElectives"
    
    # Handling questions on required electives for graduation
    elif "electives should i take" in question or "required electives" in question:
        return "MATCH (e:Elective) RETURN COUNT(e) AS TotalElectivesToGraduate"
    
    # Handling 'take in the same semester' type question
    elif "take in the same semester" in question:
        courses = question.split("take")[-1].strip().replace("?", "").split("and")
        if len(courses) == 2:
            course1, course2 = courses[0].strip(), courses[1].strip()
            return f"MATCH (c1:Course {{name: '{course1}'}}), (c2:Course {{name: '{course2}'}}) RETURN NOT EXISTS((c1)-[:PRE_REQUISITE_FOR]->(c2)) AND NOT EXISTS((c2)-[:PRE_REQUISITE_FOR]->(c1)) AS CanTakeTogether"
    
    # Handle questions about courses using regex (e.g., 'math112 course', 'csci208 course', etc.)
    elif re.search(r'course', question):
        course_name = re.sub(r'\bcourse\b', '', question).strip().replace("?", "")
        
        # If course contains multiple words, try to handle complex cases
        if '"' in course_name:
            course_name = re.sub(r'"', '', course_name).strip()  # Remove quotes from complex course names like "design and analysis of algorithms"
        
        return f"MATCH (c:Course {{name: '{course_name}'}})-[:UNLOCKS]->(unlocked:Course) RETURN unlocked.name AS UnlockedCourses"
    
    # Handle general 'track' questions like "any course in the big data track"
    elif "track" in question:
        track_name = re.search(r'in the (.*?) track', question)
        if track_name:
            track_name = track_name.group(1).strip()
            return f"MATCH (c:Course)-[:BELONGS_TO]->(t:Track {{name: '{track_name}'}}) RETURN c.name AS CoursesInTrack"
    
    # Handle questions related to courses like "any course in the data mining track"
    elif "any course" in question:
        course_name = re.sub(r'\bcourse\b', '', question).strip().replace("?", "")
        return f"MATCH (c:Course {{name: '{course_name}'}})-[:PRE_REQUISITE_FOR]->(prereq:Course) RETURN prereq.name AS Prerequisite"

    # Handle questions like "how many electives" or "how many technical electives"
    elif re.search(r'how many', question) or re.search(r'count', question):
        if "technical electives" in question:
            return "MATCH (e:Elective {type: 'technical'}) RETURN COUNT(e) AS TotalTechnicalElectives"
        return "MATCH (e:Elective) RETURN COUNT(e) AS TotalElectives"
    
    # Catch-all for other courses or ambiguous questions
    elif re.search(r'\bcsci208\b', question) or re.search(r'\bmath112\b', question):
        course_name = re.sub(r'\bcourse\b', '', question).strip().replace("?", "")
        return f"MATCH (c:Course {{name: '{course_name}'}})-[:UNLOCKS]->(unlocked:Course) RETURN unlocked.name AS UnlockedCourses"

    else:
        return "Invalid question format"

questions = [
    "What is the prerequisite for math112?",
    "What are the courses unlocked after completing math112?",
    "How many electives should I take to graduate?",
    "What is the prerequisite for csci208 course?",
    "What are the technical electives?",
    "What are the courses math112?",
    "Can I take Discrete Math and CSCI208 in the same semester?",
    "What are required electives for the computer science track?",
    "How many technical electives are required?",
    "Are there any mandatory courses for graduation?",
    "What electives should I take to graduate?",
    "What courses are required for graduation?",
    "How many core courses are in the program?",
    "Can I take CSCI208 and Discrete Math in the same semester?",
    "What are the prerequisites for CSCI207?",
    "What are the courses in the big data track?",
    "What courses are in the database systems track?",
    'The "design and analysis of algorithms" course requirements?',
    'Is the "data mining" course available in the big data track?'
]

# Generate Cypher queries for all questions
cypher_queries = [generate_cypher(q) for q in questions]


for i, query in enumerate(cypher_queries, start=1):
    print(f"Query {i}: {query}")
```
